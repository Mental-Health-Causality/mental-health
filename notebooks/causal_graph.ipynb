```python
import pandas as pd
import numpy as np
#import bnlearn as bn
#from pgmpy.models import MarkovNetwork
#G = MarkovNetwork()
from pycit.markov_blanket import MarkovBlanket
```

```python
df = pd.read_csv('../data/final.csv', sep=';')
```

```python
df_suic = df[['Suicidio','sexo', 'Estado_civil', 'Tipo_Resid','idade',
                   'Alcoolatra', 'Droga', 'Suic_familia', 'Dep_familia',
                   'Alc_familia', 'Drog_familia',
                   'Neuro',
                   'psiquiatrica', 'Anos educacao formal', 'Capaz de desfrutar das coisas',
                   'Impacto de sua familia e amigos',
                   'Capaz de tomar decisões importantes', 'Estudante',
                   'Insonia',
                   'Deprimido', 'Ansiedade',
                   'Perda de insights', 'Apetite', 'Perda de peso', 'Ansiedade somática',
                   'Hipocondriase', 'Sentimentos_culpa', 
                   'Trabalho e interesses', 'Energia', 'Lentidao pensamento e fala',
                   'Agitação', 'Libido', 'Pontuação total', 'TOC']]
```

```python
df_suic['sexo'] = np.where(df_suic['sexo'] == 'F', 1, 0)

df_suic = df_suic.dropna()
df_suic.Suicidio = np.where(df_suic.Suicidio<2,0,1)
```

```python
Y = df_suic['Suicidio']
X = df_suic.drop('Suicidio', axis = 1)
X.Estado_civil = X.Estado_civil.astype(int)
X.Tipo_Resid = X.Tipo_Resid.astype(int)
X.idade = X.idade.astype(int)
X['Anos educacao formal'] = X['Anos educacao formal'].astype(int)
X['Capaz de desfrutar das coisas'] = X['Capaz de desfrutar das coisas'].astype(int)
X['Impacto de sua familia e amigos'] = X['Impacto de sua familia e amigos'].astype(int)
X['Capaz de tomar decisões importantes'] = X['Capaz de tomar decisões importantes'].astype(int)
X['Estudante'] = X['Estudante'].astype(int)
X['Insonia'] = X['Insonia'].astype(int)
X['Deprimido'] = X['Deprimido'].astype(int)
X['Ansiedade'] = X['Ansiedade'].astype(int)
X['Perda de insights'] = X['Perda de insights'].astype(int)
X['Apetite'] = X['Apetite'].astype(int)
X['Perda de peso'] = X['Perda de peso'].astype(int)
X['Ansiedade somática'] = X['Ansiedade somática'].astype(int)
X['Hipocondriase'] = X['Hipocondriase'].astype(int)
X['Sentimentos_culpa'] = X['Sentimentos_culpa'].astype(int)
X['Trabalho e interesses'] = X['Trabalho e interesses'].astype(int)
X['Energia'] = X['Energia'].astype(int)
X['Lentidao pensamento e fala'] = X['Lentidao pensamento e fala'].astype(int)
X['Agitação'] = X['Agitação'].astype(int)
X['Libido'] = X['Libido'].astype(int)
X['Pontuação total'] = X['Pontuação total'].astype(int)
X['TOC'] = X['TOC'].astype(int)
```

```python
X
```

```python
Xd3 = []
for x in X.values:
    Xd3.append([x.tolist()])
```

```python
# specify CI test configuration
cit_funcs = {
    'it_args': {
        'test_args': {
            'statistic': 'ksg_mi',
            'n_jobs': 1
        }
    },
    'cit_args': {
        'test_args': {
            'statistic': 'ksg_cmi',
            'n_jobs': 1
        }
    }
}
```

```python
# find Markov blanket of Y. x_data contains data from predictor variables, X_1,...,X_m
mb = MarkovBlanket(np.asarray(Xd3), Y.values, cit_funcs)
```

```python
markov_blanket = mb.find_markov_blanket(max_conditioning = 2)
```
